```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file
file_path = "timing_Mp.xlsx"  # Update the path if needed
df = pd.read_excel(file_path)

# Plot M vs A and B
plt.figure(figsize=(8, 6))
plt.plot(df["M"], df["A"], label="p=0.2", marker="o")
plt.plot(df["M"], df["B"], label="p=0.4", marker="s")
plt.plot(df["M"], df["C"], label="p=0.6", marker="o")
plt.plot(df["M"], df["D"], label="p=0.8", marker="s")

# Labels and title
plt.xlabel("M")
plt.ylabel("Number of steps")
plt.title("Convergence of steps")
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel file
file_path = "timing_mp.xlsx"  # Update the path if needed
df = pd.read_excel(file_path)

# Calculate percentage change for A and B
df["A_pct_change"] = df["A"].pct_change() * 100
df["B_pct_change"] = df["B"].pct_change() * 100
df["C_pct_change"] = df["C"].pct_change() * 100
df["D_pct_change"] = df["D"].pct_change() * 100

# Plot M vs percentage change of A and B
plt.figure(figsize=(8, 6))
plt.plot(df["M"], df["A_pct_change"], label="p=0.2", marker="o")
plt.plot(df["M"], df["B_pct_change"], label="p=0.4", marker="s")
plt.plot(df["M"], df["C_pct_change"], label="p=0.6", marker="o")
plt.plot(df["M"], df["D_pct_change"], label="p=0.8", marker="s")

# Labels and title
plt.xlabel("M")
plt.ylabel("Percentage Change (%)")
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
